```python
import numpy as np
import matplotlib.pyplot as plt

def visualize_matrix(matrix_path: str, add_grid=False, dynamic_size=True, savefig_path=None):
    with open(matrix_path, 'r') as file:
        rows, cols = map(int, file.readline().split())
        matrix = np.zeros((rows, cols), dtype=int)
        for i in range(rows):
            row_data = file.readline().strip()
            matrix[i] = list(map(int, list(row_data)))

    if dynamic_size:
        figsize = (cols / 10, rows / 10)
        linewidth = min(rows, cols) / 200
    else:
        figsize = (10, 10)
        linewidth = 1

    plt.figure(figsize=figsize)
    plt.imshow(matrix, cmap='binary', interpolation='nearest', vmin=0, vmax=1)

    if add_grid:
        plt.grid(False)
        plt.gca().set_xticks(np.arange(-.5, cols, 1), minor=True)
        plt.gca().set_yticks(np.arange(-.5, rows, 1), minor=True)
        plt.gca().grid(which="minor", color="black", linestyle='-', linewidth=linewidth)

    plt.xticks([])
    plt.yticks([])
    if savefig_path is not None:
        plt.savefig(savefig_path, bbox_inches='tight')
    plt.show()
```

```python
visualize_matrix('../output/iteration_5000_10x10.txt', dynamic_size=True, savefig_path='../output/iteration_5000_10x10.png')
```

```python
visualize_matrix('../output/iteration_5000_10x10.txt', add_grid=True, dynamic_size=True, savefig_path='../output/iteration_5000_10x10.png')
```

```python
visualize_matrix('../output/iteration_5000_10x10.txt', add_grid=True, dynamic_size=False)
```

```python
visualize_matrix('../output/iteration_5000.txt', add_grid=False, dynamic_size=True, savefig_path='../output/iteration_5000.png')
```
